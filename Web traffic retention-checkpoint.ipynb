```python
import pandas as pd
import numpy as np

# Create a sample dataset of web interactions
data = {
    'session_id': [101, 101, 102, 103, 103, 103, 104, 105, 105, 106],
    'user_id': ['U1', 'U1', 'U2', 'U3', 'U3', 'U3', 'U4', 'U5', 'U5', 'U6'],
    'page_url': ['/home', '/products', '/home', '/home', '/products', '/cart', '/home', '/home', '/about', '/home'],
    'traffic_source': ['Organic', 'Organic', 'Social', 'Direct', 'Direct', 'Direct', 'Social', 'Paid', 'Paid', 'Organic'],
    'timestamp': pd.to_datetime(['2026-03-29 10:00', '2026-03-29 10:05', '2026-03-29 10:10', 
                                 '2026-03-29 10:15', '2026-03-29 10:20', '2026-03-29 10:25',
                                 '2026-03-29 10:30', '2026-03-29 10:35', '2026-03-29 10:40', 
                                 '2026-03-29 10:45'])
}

df = pd.DataFrame(data)
print("--- Raw Web Traffic Data ---")
print(df.head())
```

```python
# 1. Top Pages by Views
top_pages = df['page_url'].value_counts()

# 2. Traffic Source Distribution
source_distribution = df['traffic_source'].value_counts()

print("\n--- Top Pages ---")
print(top_pages)
print("\n--- Traffic Sources ---")
print(source_distribution)
```

```python
# Group by session_id to count how many pages were viewed in each session
session_counts = df.groupby('session_id')['page_url'].count()

total_sessions = len(session_counts)
bounced_sessions = len(session_counts[session_counts == 1])

bounce_rate = (bounced_sessions / total_sessions) * 100

print(f"\nTotal Sessions: {total_sessions}")
print(f"Bounced Sessions: {bounced_sessions}")
print(f"Overall Bounce Rate: {bounce_rate:.2f}%")
```

```python
# Define our funnel steps
funnel_steps = ['/home', '/products', '/cart']

# Count how many unique users reached each step
funnel_counts = []
for step in funnel_steps:
    count = df[df['page_url'] == step]['user_id'].nunique()
    funnel_counts.append(count)

funnel_df = pd.DataFrame({'Step': funnel_steps, 'Users': funnel_counts})

# Calculate drop-off rate
funnel_df['Retention %'] = (funnel_df['Users'] / funnel_df['Users'].iloc[0]) * 100

print("\n--- Funnel Analysis ---")
print(funnel_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. SETUP & DATA CLEANING
# We create/load data with Session IDs, URLs, and Sources
np.random.seed(42)
data = {
    'session_id': np.repeat(np.arange(100, 200), 3),
    'page_url': np.random.choice(['/home', '/products', '/cart', '/checkout', '/about', '/contact'], 300, p=[0.4, 0.25, 0.15, 0.05, 0.1, 0.05]),
    'traffic_source': np.random.choice(['Organic', 'Social', 'Direct', 'Paid'], 300),
    'timestamp': pd.date_range(start='2026-03-29', periods=300, freq='15min')
}
df = pd.DataFrame(data)
sns.set_theme(style="whitegrid")

# 2. KPI CALCULATION: BOUNCE RATE
# A bounce is a session with only 1 page view
session_counts = df.groupby('session_id')['page_url'].count()
bounce_rate = (len(session_counts[session_counts == 1]) / len(session_counts)) * 100
print(f"Project KPI - Bounce Rate: {bounce_rate:.2f}%")

# 3. VISUALIZATION 1: TOP PAGES (Popularity Analysis)
top_pages = df['page_url'].value_counts().sort_values(ascending=False).reset_index()
top_pages.columns = ['Page URL', 'Views']
plt.clf()
sns.barplot(data=top_pages, x='Views', y='Page URL', hue='Page URL', palette='viridis')
plt.title('Top Visited Pages (User Interest)')
plt.tight_layout()
plt.savefig('top_pages.png')

# 4. VISUALIZATION 2: TRAFFIC SOURCES (Acquisition)
source_counts = df['traffic_source'].value_counts()
plt.clf()
plt.pie(source_counts, labels=source_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Traffic Source Distribution')
plt.savefig('traffic_sources.png')

# 5. VISUALIZATION 3: CONVERSION FUNNEL (User Journey)
# Tracks users from Home -> Products -> Cart -> Checkout
funnel_order = ['/home', '/products', '/cart', '/checkout']
funnel_counts = [df[df['page_url'] == step]['session_id'].nunique() for step in funnel_order]
funnel_df = pd.DataFrame({'Step': funnel_order, 'Unique Sessions': funnel_counts})
plt.clf()
sns.barplot(data=funnel_df, x='Unique Sessions', y='Step', hue='Step', palette='magma')
plt.title('Website Conversion Funnel (Drop-off Analysis)')
plt.tight_layout()
plt.savefig('conversion_funnel.png')

# 6. VISUALIZATION 4: TRAFFIC TRENDS (Engagement Pulse)
df['hour'] = df['timestamp'].dt.hour
hourly_traffic = df.groupby('hour').size().reset_index(name='Hits')
plt.clf()
sns.lineplot(data=hourly_traffic, x='hour', y='Hits', marker='o', color='teal')
plt.title('Hourly Engagement Trends')
plt.xlabel('Hour of Day')
plt.ylabel('Total Hits')
plt.tight_layout()
plt.savefig('traffic_over_time.png')

# Export for report
df.to_csv('web_traffic_analytics_final.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. GENERATE SAMPLE DATA (Matches the project brief)
np.random.seed(42)
data = {
    'session_id': np.repeat(np.arange(100, 250), 3),
    'page_url': np.random.choice(['/home', '/products', '/cart', '/checkout', '/about'], 450, p=[0.4, 0.25, 0.15, 0.1, 0.1]),
    'traffic_source': np.random.choice(['Organic', 'Social', 'Direct', 'Paid'], 450),
    'timestamp': pd.date_range(start='2026-03-29', periods=450, freq='10min')
}
df = pd.DataFrame(data)
df['hour'] = df['timestamp'].dt.hour

# 2. SET UP THE DASHBOARD (2 Rows, 2 Columns)
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Project 3: Web Traffic Analytics Dashboard', fontsize=22, fontweight='bold')

# --- CHART 1: TOP PAGES (Top Left) ---
top_pages = df['page_url'].value_counts().reset_index()
top_pages.columns = ['Page', 'Views']
sns.barplot(ax=axes[0, 0], data=top_pages, x='Views', y='Page', hue='Page', palette='viridis', legend=False)
axes[0, 0].set_title('Top Visited Pages (Popularity)', fontsize=14)

# --- CHART 2: TRAFFIC SOURCES (Top Right) ---
source_counts = df['traffic_source'].value_counts()
axes[0, 1].pie(source_counts, labels=source_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
axes[0, 1].set_title('Traffic Source Distribution (Acquisition)', fontsize=14)

# --- CHART 3: CONVERSION FUNNEL (Bottom Left) ---
# Logic: Tracking users through the buying journey
funnel_order = ['/home', '/products', '/cart', '/checkout']
funnel_counts = [df[df['page_url'] == step]['session_id'].nunique() for step in funnel_order]
sns.barplot(ax=axes[1, 0], x=funnel_counts, y=funnel_order, hue=funnel_order, palette='magma', legend=False)
axes[1, 0].set_title('User Conversion Funnel (Drop-off)', fontsize=14)
axes[1, 0].set_xlabel('Unique Sessions')

# --- CHART 4: HOURLY TRENDS (Bottom Right) ---
hourly_traffic = df.groupby('hour').size().reset_index(name='Hits')
sns.lineplot(ax=axes[1, 1], data=hourly_traffic, x='hour', y='Hits', marker='o', color='teal', linewidth=2.5)
axes[1, 1].set_title('Hourly Website Traffic (Engagement)', fontsize=14)
axes[1, 1].set_xlabel('Hour of Day (0-23)')

# 3. FINALIZE LAYOUT
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('web_traffic_analytics_dashboard.png', dpi=300) # Saves as a high-quality image
plt.show()

# 4. KPI PRINT-OUT (For your report text)
session_counts = df.groupby('session_id')['page_url'].count()
bounce_rate = (len(session_counts[session_counts == 1]) / len(session_counts)) * 100
print(f"--- Key Performance Indicators ---")
print(f"Overall Bounce Rate: {bounce_rate:.2f}%")
print(f"Total Unique Sessions: {df['session_id'].nunique()}")
```
